```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# dataset 1

## Training set Loss

```python
cnn_train_loss = pd.read_csv('data1/cnn_train_loss.csv')
resnet_train_loss = pd.read_csv('data1/resnet_train_loss.csv')
resnet_ft_train_loss = pd.read_csv('data1/resnet_ft_train_loss.csv')
vgg_ft_train_loss = pd.read_csv('data1/vgg_ft_train_loss.csv')
```

```python
plt.plot(cnn_train_loss['Step'], cnn_train_loss['Value'], label = "CNN")
plt.plot(resnet_train_loss['Step'], resnet_train_loss['Value'], label = "Resnet")
plt.plot(resnet_ft_train_loss['Step'], resnet_ft_train_loss['Value'], label = "Resnet+ft")
plt.plot(vgg_ft_train_loss['Step'], vgg_ft_train_loss['Value'], label = "VGG+ft")

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Val set Loss

```python
cnn_val_loss = pd.read_csv('data1/cnn_val_loss.csv')
resnet_val_loss = pd.read_csv('data1/resnet_val_loss.csv')
resnet_ft_val_loss = pd.read_csv('data1/resnet_ft_val_loss.csv')
vgg_ft_val_loss = pd.read_csv('data1/vgg_ft_val_loss.csv')
```

```python
plt.plot(cnn_val_loss['Step'], cnn_val_loss['Value'], label = "CNN")
plt.plot(resnet_val_loss['Step'], resnet_val_loss['Value'], label = "Resnet")
plt.plot(resnet_ft_val_loss['Step'], resnet_ft_val_loss['Value'], label = "Resnet+ft")
plt.plot(vgg_ft_val_loss['Step'], vgg_ft_val_loss['Value'], label = "VGG+ft")

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## LR

```python
cnn_lr = pd.read_csv('data1/cnn_lr.csv')
resnet_lr = pd.read_csv('data1/resnet_lr.csv')
resnet_ft_lr = pd.read_csv('data1/resnet_ft_lr.csv')
vgg_ft_lr = pd.read_csv('data1/vgg_ft_lr.csv')
```

```python
plt.plot(cnn_lr['Step'], cnn_lr['Value'], label = "CNN")
plt.plot(resnet_lr['Step'], resnet_lr['Value'], label = "Resnet")
plt.plot(resnet_ft_lr['Step'], resnet_ft_lr['Value'], label = "Resnet+ft")
plt.plot(vgg_ft_lr['Step'], vgg_ft_lr['Value'], label = "VGG+ft")

plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.legend()
plt.show()
```
